# Chaos Game and Sierpinski Triangle
Chaos game is a method of creating a fractal,using polygon and an intial point 
selected at random inside it

The fractal is created by iteratively creating a sequence of points , starting with intial
random point in which each point in the sequence is a given fraction of the distance between 
previous point and one of the vertices of the polygon.
Sierpinski Triangle can be generated by using regular triangle as polygon and factor of 0.5.

```python
import random
import matplotlib.pyplot as plt
x1 = 2;
x2 = 5
x3 = 8
y1 = 1
y2 = 6.196
y3 = 1;
x_values = [x1,x2,x3];
y_values = [y1,y2,y3];
plt.scatter(x_values,y_values,marker = ".");
plt.annotate("A",(x1,y1));
plt.annotate("B",(x2,y2));
plt.annotate("C",(x3,y3));

start_point_x = random.uniform(x1,x2);
start_point_y = random.uniform(y1,y3);

next_point_x = start_point_x;
next_point_y = start_point_y;
for i in range(1,5000):
    decide_point = random.randint(1,3);
    
    if decide_point == 1:
        next_point_x = abs(next_point_x+x1)/2;
        next_point_y = abs(next_point_y+y1)/2;
        
    else:
        if decide_point ==2:
            next_point_x = abs(next_point_x+x2)/2;
            next_point_y = abs(next_point_y+y2)/2;
            
        else:
            next_point_x = abs(next_point_x+x3)/2;
            next_point_y = abs(next_point_y+y3)/2;
            
    plt.scatter(next_point_x,next_point_y,marker=".",c ="red");
plt.title("Sierpinski Triangle")
```

**Animation of Sierpinski Triangle using turtle**

```python
from turtle import *
right(30)
pu()
ht()
shape("triangle")
speed(75)
def triangle(a,b):
    if b==0:
        stamp()
    else:
        for i in range(0,3):
            forward(a)
            triangle(a/2,b-1)
            backward(a)
            left(120)
triangle(100,4)
```

**Fractals in Square**
My idea of generating fractals in square using Sierpinski triangle involves rotating 
the generated triangle 4 times. But in order to generate square as base polygon from 
triangle, the triangle must be an isosceles right angled triangle. Hence I have considered
the coordinates of triangle accordingly.

```python
import random 
import matplotlib.pyplot as plt;
x1 = 0;
x2 = 10;
x3 = 5;
x_values = [x1,x2,x3];
y1 = 0;
y2 = 0;
y3 = 5;
y_values = [y1,y2,y3];
plt.xlim(-1,11);
plt.ylim(-1,11);
plt.scatter(x_values,y_values,marker = ".",c = "red");
dis1 = pow((x1-x2),2);
dis2 = pow((y1-y2),2);
dis3 = dis1 + dis2;
dis = pow(dis3,0.5);
plt.annotate("C",(x3,y3));


def sierpinski_triangle(x1,y1,x2,y2,x3,y3):
    
    start_point_x = random.uniform(x1,x2);
    start_point_y = random.uniform(y1,y3);
    start_point_y = (y1+y2+y3)/3;
    plt.scatter(start_point_x,start_point_y,marker = ".",c = "green");
    next_point_x = start_point_x;
    next_point_y = start_point_y;
    for i in range(1,5000,2):
        decide_point = random.randint(1,3);
        
        if decide_point == 1:
            next_point_x = abs(next_point_x+x1)/2;
            next_point_y = abs(next_point_y+y1)/2;
            
        else:
            if decide_point ==2:
                next_point_x = abs(next_point_x+x2)/2;
                next_point_y = abs(next_point_y+y2)/2;
               
            else:
                next_point_x = abs(next_point_x+x3)/2;
                next_point_y = abs(next_point_y+y3)/2;
               
        plt.scatter(next_point_x,next_point_y,marker=".",c ="green");
sierpinski_triangle(x1,y1,x2,y2,x3,y3);
sierpinski_triangle(x1,y1+dis,x2-dis,y2,x3,y3);
sierpinski_triangle(x2,y2,x2,y2+dis,x3,y3);
sierpinski_triangle(x1,y1+dis,x2,y2+dis,x3,y3);
plt.title("Fractals in square")
```
